# Feature Scaling 

```python
#miport dataset 
import pandas as pd
insurance=pd.read_csv('insurance.csv')
supershop=pd.read_csv('supershop.csv')
worldometer=pd.read_csv('worldometer_data.csv')
```

```python
insurance.head()
```

```python
insurance.isnull().sum()
```

```python
insurance.shape
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
from sklearn.preprocessing import MaxAbsScaler 
```

```python
mb=MaxAbsScaler()
```

```python
insurance['charges']=mb.fit_transform(insurance[['charges']])
```

```python
insurance.head()
```

```python
#Another way
```

```python
for dataset in supershop.columns:
    if is_numeric_dtype(supershop[dataset]):
        supershop[dataset]=mb.fit_transform(supershop[[dataset]])
    else:
        continue
```

```python
supershop.head()
```

```python
supershop.isnull().sum()
```

```python
#supershop.interpolate(inplace=True)
supershop.fillna(method='bfill',inplace=True)
```

```python
supershop.isnull().sum()
```

```python
supershop.head()
```

# Encoding

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
insurance['region']=le.fit_transform(insurance['region'])
```

```python
insurance.head()
```

```python
#Another way
```

```python
for dataset in supershop.columns:
    if is_numeric_dtype(supershop[dataset]):
        continue
    else:
        supershop[dataset]=le.fit_transform(supershop[dataset])
```

# Ordinal-Encoder

```python
 from sklearn.preprocessing import OrdinalEncoder
```

```python
smok=['yes','no']
```

```python
oe=OrdinalEncoder(categories=[smok])
```

```python
oe
```

```python
insurance1=insurance.copy()
```

```python
insuran=oe.fit_transform(insurance[['smoker']])
```

```python
insu=pd.DataFrame(insuran,columns=['smokerN'])
```

```python
insurance=pd.concat([insurance,insu],axis=1)
```

```python
insurance.drop('smoker',axis=1,inplace=True)
```

```python
insurance.head()
```

```python
#Another way
column=['smoker']
```

```python
insurance1.head(2)
```

```python
for i in column:
    unique=insurance1[i].unique()
    insurance1[i]=OrdinalEncoder(categories=[unique]).fit_transform(insurance1[[i]])
    
```

```python
insurance1.head()
```
