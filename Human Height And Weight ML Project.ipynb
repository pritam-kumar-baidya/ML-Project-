```python
mp.figure(figsize=(18,15))
mp.scatter(weight_height['Height'],weight_height['Weight'],marker='*',c="tomato") 
mp.xlabel('Height')
mp.ylabel('Weight')
mp.title('Human Height And Weight',size=40)
mp.show()
```

```python
#Divide by variable
x=weight_height[['Height']]
y=weight_height[['Weight']]
```

```python
#import ML Algorithm
from sklearn.model_selection import train_test_split as ts
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#data testing and training
xtrain,xtest,ytrain,ytest=ts(x,y,train_size=.70,random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
lr=LinearRegression()
```

```python
#model fit
lr.fit(xtrain,ytrain)
```

```python
#predicted value
lr.predict([[73.847017]])
```

```python
ypred=lr.predict(xtest)
```

```python
mean_squared_error(ytest,ypred)
```

```python
mean_absolute_error(ytest,ypred)
```

```python
lr.score(xtest,ytest)
```

```python
lr.score(xtrain,ytrain)
```

```python
lr.score(weight_height[['Height']],weight_height[['Weight']])
```
