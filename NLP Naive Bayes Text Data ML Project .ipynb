# Text-Preprocessing

```python
def clean_text_data(Message):
    a =[char for char in Message if char not in string.punctuation]
    a =''.join(a)
    
    c = [word for word in a.split() if word.lower not in stopwords.words()]
    return c
            
```

```python
df.Message.apply(clean_text_data)
```

```python
df.head()
```

# Feature-Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv=TfidfVectorizer(analyzer=clean_text_data).fit_transform(df['Message'])
```

# Split-the-dataset

```python
y=df[['Category']]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain, xtest, ytrain, ytest = tts(tv,y,train_size=.70,random_state=41)
```

# Mutinomial-Naive-Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(xtrain,ytrain)
```

```python
mnb.score(xtest, ytest)
```

```python
mnb.predict(xtest)
```

# Bernoulli-Naive-Bayes

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(xtrain,ytrain)
```

```python
bnb.score(xtest, ytest)
```
